#### Wiki API

```python
import requests

urls = ["https://www.wikiwand.com/en/Python_(programming_language)", # Page 1
        "https://en.wikipedia.org/wiki/Artificial_intelligence",      # Page 2
       "https://en.wikipedia.org/wiki/Quantum_computing",            # Page 3
    "https://en.wikipedia.org/wiki/Machine_learning"              # Page 4
]

for i, url in enumerate(urls):

    response = requests.get(url)
    with open(f"wikipedia_page_{i+1}.html", "w", encoding='utf-8') as f:
        contents = f.write(response.text)


print (contents)
```

```python

#Create an empty graph with no nodes and no edges.
import networkx as nx
G = nx.Graph()
```

```python
G.add_nodes_from([1,2,3])

G.add_nodes_from([
    (4, {"color": "red"}),
    (5, {"color": "green"}),
])

G.add_edge(1, 2)
e = (2, 3)
G.add_edge(*e)  # unpack edge tuple*
print(G)
```

```python
import matplotlib.pyplot as plt
G = nx.petersen_graph()
subax1 = plt.subplot(121)
nx.draw(G, with_labels=True, font_weight='bold')
subax2 = plt.subplot(122)
nx.draw_shell(G, nlist=[range(5, 10), range(5)], with_labels=True, font_weight='bold')
```
